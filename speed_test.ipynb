```
from timeit import default_timer as dt

class SpeedTest:
  def __enter__(self):
    self.start = dt()
    return self

  def __exit__(self, type, *args):
    end = dt()
    print(end - self.start)

with SpeedTest():
  for i in range(10000):
    for g in range(50000):
      z = i*g
print(z)
```
